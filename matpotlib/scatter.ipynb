```python
import matplotlib.pyplot as plt
```

```python
# Example: Hours of Study vs Exam Score (20 students)
hours_studied = [1, 2, 2.5, 3, 3.5, 
                 4, 4.5, 5, 5.5, 6, 
                 6.5, 7, 7.5, 8, 8.5, 
                 9, 9.5, 10, 10.5, 11]

exam_scores = [40, 35, 48, 52, 55, 
               60, 62, 65, 58, 72, 
               74, 78, 80, 93, 85, 
               88, 90, 82, 94, 96]
plt.style.use("classic")
size = [i-20 for i in exam_scores]
colors = ["red" for i in exam_scores]
plt.scatter(hours_studied, exam_scores, c = colors, s = size)
plt.grid()
```

```python
size = [i-20 for i in exam_scores]
plt.scatter(hours_studied, exam_scores,c = exam_scores, cmap = 'Reds',s = size)

for i in range(len(exam_scores)):
    plt.annotate(f's{i}', (hours_studied[i] + 0.4, exam_scores[i]))
plt.colorbar()
plt.grid()
```

```python
hours_studied2 = [1, 2, 2.5, 3, 3.5,
                  4, 4.5, 5, 5.5, 6,
                  6.5, 7, 7.5, 8, 8.5,
                  9, 9.5, 10, 10.5, 11]

exam_scores2 = [35, 40, 42, 46, 50,
                53, 55, 58, 60, 63,
                65, 68, 70, 72, 74,
                76, 78, 80, 81, 83]
size2 = [i for i in exam_scores2]
plt.scatter(hours_studied, exam_scores, c = exam_scores, s = size, cmap = "Blues", label = "sec A")
plt.scatter(hours_studied2, exam_scores2, c = exam_scores2, s = size2, cmap = "Reds", label = "sec B")
plt.colorbar()
plt.xlabel("hours studied")
plt.ylabel("marks")
plt.legend()
plt.title("sec 'A' vs 'B'")
```
