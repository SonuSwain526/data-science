```python
years_rohit = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]
rohit = [532, 102, 504, 611, 168, 1196, 578, 815, 564, 1293, 1030, 1490]
years = [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

virat = [159, 325, 995, 1381, 1026, 1268, 1054, 623, 739, 1460, 1202, 1377, 431, 129, 302, 1377]

plt.plot(years_rohit, rohit, 'r3--', label = "rohit")
plt.plot(years, virat,'go-.', label = "kohli")
plt.legend()
plt.xlabel("year")
plt.ylabel("runs")
plt.title("kohli vs rohit in odis")
```
