```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Book1.csv")
df
```

```python
days = np.array(df['days'])
book = np.array(df['book'])
coding = np.array(df['coding'])
projects = np.array(df['projects'])
sleeping = np.array(df['sleeping'])

labels = [book, coding, projects, sleeping]
plt.stackplot(days, book, coding, projects, sleeping, labels=labels, alpha = 0.7)
plt.legend(labels)
plt.xlabel("days")
plt.ylabel("hours")
plt.title("my weekly timetable")
plt.grid()
plt.figure(figsize=(6,10))
```
