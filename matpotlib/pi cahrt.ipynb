```python
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-colorblind")
```

```python
players = ["pradeep", "manji", "pawan", "Shardul", "fazal"]
raids = [2000, 1500, 1700, 1000-500, 200] 

plt.pie(raids, labels=players, colors=["#aac3f6","#c7d7a1","#eb86ef", "#b15a5a"])
plt.title("top player")
```

```python

plt.pie(
    raids,
    labels=players,
    colors = ["#aac3f6","#aaab34","#c7d7a1","#eb86ef", "#b15a5a"],
    wedgeprops={"edgecolor": "black", "linewidth" : 0.5, "linestyle": '-.'},
    startangle=100,
    explode=[0,0,0,0.17,0],
    shadow=True,
    autopct="%1.2f%%"
)
```
