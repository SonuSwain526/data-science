```python
sehwag = [0, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 0]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010] 
kohli =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500] 
me =  [900, 700, 1180, 1142, 1200, 1650, 1200, 1200, 0, 124] 

plt.plot(years, sehwag, label = "Sehwag bhai")
plt.plot(years, kohli, label = "kohli bhai")
plt.plot(years, me, label = "Gourab bhai")

plt.xlabel("years")
plt.ylabel("runs")
plt.title("kohli vs sehwag")
plt.legend()
```

```python
plt.plot(years, sehwag,'m1-.', label = "Sehwag bhai")
plt.plot(years, kohli, 'b3--', label = "kohli bhai")
plt.plot(years, me, 'yp-',label = "Gourab bhai")

plt.legend()
```
