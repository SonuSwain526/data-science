```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
sehwag = [100, 300, 800, 1200, 1500, 1700, 1600, 1400, 1000, 100]
years = [1990, 1992, 1994, 1996, 1998, 2000, 2003, 2005, 2007, 2010] 
kohli =  [500, 700, 1100, 1500, 1800, 1200, 1700, 1300, 900, 1500] 

plt.bar(years, kohli,)
# plt.barh(years, kohli,)
# plt.barh(kohli, years)
for i in range(len(years)):
    plt.text(years[i], kohli[i] + 30, kohli[i], ha = "center")
plt.xlabel("years")
plt.ylabel("runs")
plt.title("kohi runs in odi")
```

```python
x = np.arange(len(years))
x
```

```python
width = 0.3
plt.bar(x - 0.3, sehwag, width=width, label = "sehwag sir")
plt.bar(x, kohli, width=width, label = "kohlis sir")
plt.legend()
plt.xlabel("years")
plt.ylabel("runs")
plt.title("sachi vs virat in bar form")
plt.xticks(x, years)

```
