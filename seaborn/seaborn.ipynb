```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset("car_crashes")
sns.lineplot(x="total", y="alcohol", data=data, color = "red", label = "beudaa")
sns.lineplot(x="total", y="speeding", data=data, color = "yellow", label = "loudubaaz")
plt.title("Accident compare betwen alcoho and speed")
plt.legend()
plt.tight_layout()
plt.grid()
```

```python
sns.barplot(x="total", y="no_previous", data=data)
plt.tight_layout()
```
