```python
floted_arr = np.array([1.1, 2.2, 3.3])
print(floted_arr.dtype)
int_arr =  np.array([1.1, 2.2, 3.3], dtype = np.int32)   
int_arr
```

Use fancy indexing to extract even numbers from  [1, 2, 3, 4, 5, 6] .

```python
arr = np.array([1, 2, 3, 4, 5, 6])
idx = arr % 2 == 0
print(arr[idx])
```
