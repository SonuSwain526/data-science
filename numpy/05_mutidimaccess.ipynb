```python
import numpy as np
```

```python
arr1d = np.array([12,43,90,2,0,])
```

```python
arr2d = np.array([[1,2,3], 
                  [12,3,44]])
arr2d
```

```python
print("rows sum 1d :", np.sum(arr1d, axis = 0))
print("rows sum 2d :", np.sum(arr2d, axis = 0))
print("rows sum 2d :", np.sum(arr2d, axis = 1))
```

```python
print(arr2d[0])
print(arr2d[:][1])
```

slicing of multid

```python
print(arr2d[0:,1:])
```

```python
arr3d = np.array([
                   [[12,32,3],
                   [10,0,22]], 
                    
                   [[90,9,3], 
                   [40,50,60]]
                 ])
arr3d.ndim
```

indexing in 3d array

```python
arr3d[1][1][1]
```

```python
arr3d[:, 1, :]
arr3d[:, 1, 1]
```

```python
print("arr axis0 sum :", np.sum(arr3d, axis = 0))
print("arr axis1 sum :", np.sum(arr3d, axis = 1))
print("arr axis2 sum :", np.sum(arr3d, axis = 2))
```

```python
arr2d[:,0] = 0
print(arr2d)
arr3d[:, 1, :] = 3
print(arr3d)
```
