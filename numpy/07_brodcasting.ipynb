applying broadcasting in real life scenarios

```python
 data = np.array([[10, 20, 30],
 [15, 25, 35],
 [20, 30, 40],
 [25, 35, 45],
 [30, 40, 50]])

mean = np.mean(data, axis = 0)
standard_dev = np.std(data, axis = 0)

normalized_Data = (data - mean) / standard_dev

print(normalized_Data)
```
