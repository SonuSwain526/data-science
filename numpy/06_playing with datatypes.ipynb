```python
import numpy as np
```

```python
# downcasitng at means changing data types to higher to lower as they use more memory
arr_bit = np.array([1000,200000,30000], dtype = np.int64)
arr_small = arr_bit.astype('int32')
print(arr_bit.dtype)
print(arr_small.dtype)
```

```python
print(arr_bit.nbytes)
print(arr_small.nbytes)
```
