```python
import pandas as pd
```

```python
 data = {
 'Name': ['Alice', 'Bob', 'Charlie'],
 'Math': [85, 78, 92],
 'Science': [90, 82, 89],
 'English': [88, 85, 94]
 }

df = pd.DataFrame(data)
df
```

```python
df2 = df.melt(id_vars="Name", value_vars=['Math', 'Science', 'English'], var_name=['subject'],value_name='Schore').copy()
```

```python
df2
```

```python
df2.pivot(index='Name', columns='subject', values='Schore')
```
