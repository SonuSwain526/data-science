```python
import pandas as pd
```

```python
df1 = pd.DataFrame({"EmpID": [1, 2, 3],
 "Name": ["Alice", "Bob", "Charlie"],
 "DeptID": [10, 20, 30]})
df1
```

```python
df2 = pd.DataFrame({"DeptID": [10, 20, 40],
 "DeptName": ["HR", "Engineering", "Marketing"]})
df2
```

```python
pd.merge(df1, df2, on = "DeptID")
```

```python
pd.merge(df1, df2, on = "DeptID", how = "left")
```

```python
pd.merge(df1, df2, on = "DeptID", how = "right")
```

```python
pd.merge(df1, df2, on = "DeptID", how = "outer")
```

CONCATINATION

```python
df3 = pd.DataFrame({"Name":["Alice","Bob"]})
df4 = pd.DataFrame({"Name":["Charlie","David"]})
```

```python
pd.concat([df3, df4])
```

```python
pd.concat([df3, df4], axis=1)
```
