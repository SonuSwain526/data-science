```python
import pandas as pd
import numpy as np
from sklearn import tree
```

```python
data = pd.read_excel('train.xlsx')
```

```python
data.tail()
```

```python
cls = tree.DecisionTreeClassifier()
```

```python
x = data.iloc[:, :-1]
x_formatdata = x.values
```

```python
y = data.iloc[:, -1]
y_formatdata = y.values
```

```python
model = cls.fit(x_formatdata, y_formatdata)
```

```python
testing_Data = pd.read_excel("test.xlsx")
testing_Data.info()
xtest = testing_Data.iloc[:, :-1]
xtest_format = xtest.values
ytest = testing_Data.iloc[:,-1]
ytest_format = ytest.values
model.predict(xtest_format)
ytest_format
```

```python
count = 1
for i in range(0,len(xtest)):
    if(model.predict([xtest_format[i]]) == ytest_format[i]):
        count += 1
count
```

```python
accurect = (count * 100)/len(ytest_format)
accurect
```
